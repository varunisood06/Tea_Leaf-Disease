```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

datagen = ImageDataGenerator(
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest',
    validation_split=0.2) # set validation split to 0.2

train_data = datagen.flow_from_directory('/kaggle/working/tea-disease-dataset/datasets/train',
                                         target_size=(224,224),
                                         batch_size=32,
                                         class_mode='categorical',
                                         subset='training') # set subset to 'training' for train data

val_data = datagen.flow_from_directory('/kaggle/working/tea-disease-dataset/datasets/train',
                                       target_size=(224,224),
                                       batch_size=32,
                                       class_mode='categorical',
                                       subset='validation') # set subset to 'validation' for validation data

test_data = datagen.flow_from_directory('/kaggle/working/tea-disease-dataset/datasets/test',
                                         target_size=(224,224),
                                         batch_size=32,
                                         class_mode='categorical')
```
